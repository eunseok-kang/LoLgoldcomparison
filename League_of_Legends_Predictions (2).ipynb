```python
#This defines a subset in the lol dataset that I imported. A lot of columns were negligible/not correlated with my research question, so I picked out the main 5 and defined my subset off of it. 
lol_subset = lol[['Kills', 'Deaths',
       'Assists', 'Creep Score', 'Gold Earned']]
```

```python
#this shows the top 5 rows within my subset columns
lol_subset.head()
```

```python
fig, axs = plt.subplots(ncols=5, nrows=1, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in lol_subset.items():
    sns.distplot(v, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
#This creates histograms with a line which shows the general pathing of the histogram/pattern. 
```

```python
plt.figure(figsize = (15,9))
sns.heatmap(lol_subset.corr(), annot = True, cmap = 'Blues')
plt.title("Pearson Correlation")
plt.show()
#This creates a correlation graph with a blue color scheme. The darker the color, the more positively correlated the two values are. 
#Since two columns which have the same name are the exact same, they have the maximum value of 1, and have the darkest shade of blue. 
```

```python
from sklearn import preprocessing
# Let's scale the columns before plotting them against MEDV
min_max_scaler = preprocessing.MinMaxScaler()
column_sels = ['Kills', 'Deaths',
       'Assists', 'Creep Score']
x = lol_subset.loc[:,column_sels]
y = lol_subset['Gold Earned']
x = pd.DataFrame(data=min_max_scaler.fit_transform(x), columns=column_sels)
fig, axs = plt.subplots(ncols=4, nrows=1, figsize=(20, 10))
index = 0
axs = axs.flatten()
for i, k in enumerate(column_sels):
    sns.regplot(y=y, x=x[k], ax=axs[i])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
#This creates linear regression models, which also show the standard error with diagonal lines spreading from the main linear regression line. 
#This supports the correlation graph by showing the steep positive slopes of the Kills and Creep Score graphs, indicating the same idea of high positive correlation between the two valeus and gold earned!
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 4)
```

```python
# Import library for Linear Regression
from sklearn.linear_model import LinearRegression

# Create a Linear regressor
lm = LinearRegression()

# Train the model using the training sets 
lm.fit(X_train, y_train)
```

```python
lm.intercept_
#defines my interecept
```

```python
coeffcients = pd.DataFrame([X_train.columns,lm.coef_]).T
coeffcients = coeffcients.rename(columns={0: 'Attribute', 1: 'Coefficients'})
coeffcients
#shows my coefficients for my linear regression
```

:Gold Earned=5215.201431*Kills+-172.696797*Deaths+3344.150918*Assists+9841.614094*Creep Score+4418.176592812175

```python
y_pred = lm.predict(X_train)
#defines my training dataset
```

```python
print('R^2:',metrics.r2_score(y_train, y_pred))
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_train, y_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_train, y_pred))
print('MSE:',metrics.mean_squared_error(y_train, y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
#all the values from my training dataset, including the accuracy/R^2 value of my model, and the MAE/MSE, which defines the possible errors in my model.
```

```python
y_test_pred = lm.predict(X_test)
#defines testing model
```

```python
acc_linreg = metrics.r2_score(y_test, y_test_pred)
print('R^2:', acc_linreg)
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_test_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_test, y_test_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
#all the values from my testing dataset after my model went through training through the training dataset.
#we can see that the R^2 value has increased, and the MAE/MSE/RMSE values have decreased slightly after training
#this means that the accuracy of my linear regression model has increased, and the error values have decreased after training the model from the training dataset!
```
